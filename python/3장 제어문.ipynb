```python
# 주머니에 카드가 없다면 걸어가고, 있다면 버스를 타고 가라
pocket = ['card']
if 'card' in pocket:
    print("버스를 타고 가라")
else:
    print('걸어가라')
```

```python
# 1부터 10까지의 숫자 중에서 3의 배수를 뺀 나머지 값을 출력해 보자
a = 1
while a < 10:
    a = a + 1
    if a % 3 == 0: continue
    print(a)
```

```python
# for문과 range 함수를 사용하여 1부터 100까지 더해 보자
add = 0
for i in range(1,101):
    add = add + i
print(add)
```
