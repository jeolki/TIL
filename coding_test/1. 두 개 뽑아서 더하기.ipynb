```python
numbers = [2,1,3,4,1]

def solution(numbers):
    answer = []

    for i in range(len(numbers)):
        for j in range(len(numbers)):
            if i == j:
                continue
            answer.append(numbers[i] + numbers[j])
    answer = sorted(list(set(answer)))
    
    return answer

answer = solution(numbers)
print(answer)
```
